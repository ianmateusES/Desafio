```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (20, 5)
```

```python
path = "../data/"
filename = "data_1__.csv"
file = path + str(filename)
df1 = pd.read_csv(file, delimiter=",")
```

```python
path = "../data/"
filename = "data_2__.csv"
file = path + str(filename)
df2 = pd.read_csv(file, delimiter=",")
```

```python
path = "../data/"
filename = "data_4__.csv"
file = path + str(filename)
df4 = pd.read_csv(file, delimiter=",")
```

```python
df1['data'] = pd.to_datetime(df1['data'])
```

```python
AnoMesUnico = df1['data'].dt.strftime('%Y%m').unique()

acumulos = []
for i in AnoMesUnico:
    acumulos.append(df1[df1['data'].dt.strftime('%Y%m') == (i)].sum().valor)

df1_f = pd.DataFrame({'data': AnoMesUnico,'valor': acumulos})
```

```python
df1_f.plot(x='data', y='valor')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

from sklearn import metrics
import numpy as np
```

```python
model = {}
model['LinearRegression'] = LinearRegression()
model['PolynomialFeatures'] = PolynomialFeatures()
```

```python
data201810 = df1_f['valor'].values[-1]
df1_f.drop(df1_f.index[-1], inplace=True)
```

```python
target_df1_f = [[data201810]]

X = pd.DataFrame(df1_f['data'])   #Variavel independente
y = pd.DataFrame(df1_f['valor'])  #Variavel dependente

for i in model:
    i.fit(X, y)
    pred_df1_f = (model.predict([[201810]]))
    y_test = pred_df1_f
    y_pred = target_df1_f
    print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
    print('MSE:', metrics.mean_squared_error(y_test, y_pred))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
